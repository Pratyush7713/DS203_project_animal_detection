make all the images 800*600

```python
import os
import cv2
from PIL import Image
```

```python
# --- paths ---
input_dir = "/home/pratyush/Desktop/DS_Project/DS203-2025-S1-E7-project-images/images"       # change this to your images folder
output_dir = "/home/pratyush/Desktop/DS_Project/data/preprocessed"

os.makedirs(output_dir, exist_ok=True)

TARGET_W, TARGET_H = 800, 600
TARGET_RATIO = TARGET_W / TARGET_H  # 4:3 = 1.333...

def is_approx_4_3(w, h, tol=0.05):
    """Check if the image is roughly 4:3 ratio."""
    ratio = w / h
    return abs(ratio - TARGET_RATIO) < tol

def crop_to_4_3(img):
    """Center-crop any image to 4:3 ratio."""
    h, w = img.shape[:2]
    new_w = int(h * TARGET_RATIO)
    new_h = int(w / TARGET_RATIO)

    # Decide whether to crop width or height
    if new_w <= w:
        x1 = (w - new_w) // 2
        return img[:, x1:x1 + new_w]
    else:
        y1 = (h - new_h) // 2
        return img[y1:y1 + new_h, :]

count = 0
skipped = 0

for fname in os.listdir(input_dir):
    if not fname.lower().endswith((".jpg", ".jpeg", ".png")):
        continue

    fpath = os.path.join(input_dir, fname)
    img = cv2.imread(fpath)

    if img is None:
        print(f"❌ Could not read {fname}")
        continue

    h, w = img.shape[:2]

    # Skip if either side smaller than 800×600
    if w < TARGET_W or h < TARGET_H:
        # print(fname)
        skipped += 1
        continue

    # Crop if not close to 4:3
    if not is_approx_4_3(w, h):
        img = crop_to_4_3(img)
        h, w = img.shape[:2]

    # Resize down to 800×600
    img_resized = cv2.resize(img, (TARGET_W, TARGET_H), interpolation=cv2.INTER_AREA)

    out_path = os.path.join(output_dir, fname)
    cv2.imwrite(out_path, img_resized)
    count += 1

print(f"✅ Processed {count} images and saved to {output_dir}")
print(f"⚠️ Skipped {skipped} smaller images (<800×600)")
```
